### 1. What is name of car that has the highest horsepower?  

```python
try:
    df_cars = pd.read_csv('Cars.csv')
    print("Dataset reloaded successfully for this query.")
except FileNotFoundError:
    print("Error: 'Cars.csv' not found. Please ensure the CSV file is in the same directory.")
    exit()
car_with_max_hp = df_cars.loc[df_cars['horsepower'].idxmax()]
name_of_car = car_with_max_hp['name']
print(f"\nThe car with the highest horsepower is: {name_of_car}")
```

### 2. How many cars have mpg  35?

```python
cars_with_high_mpg = df_cars[df_cars['mpg'] >= 35]
num_cars_high_mpg = cars_with_high_mpg.shape[0]
print(f"\nNumber of cars with mpg  35: {num_cars_high_mpg}")
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = df_cars[(df_cars['horsepower'] > 100) & (df_cars['weight'] < 3000)]
if not filtered_cars.empty:
    most_common_origin = filtered_cars['origin'].value_counts().idxmax()
    print(f"\nThe most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
else:
    print("\nNo cars found matching the specified criteria (horsepower > 100 and weight < 3000).")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japanese_cars = df_cars[df_cars['origin'] == 'Japan']
if not japanese_cars.empty:
    mean_accel_japan = japanese_cars['acceleration'].mean()
    rounded_mean_accel = round(mean_accel_japan, 2)
    print(f"\nThe mean acceleration of cars from Japan is: {rounded_mean_accel}")
else:
    print("\nNo cars found from Japan in the dataset to calculate mean acceleration.")
```

### 5. Which year had the highest average mpg?

```python
average_mpg_by_year = df_cars.groupby('model_year')['mpg'].mean()
year_highest_avg_mpg = average_mpg_by_year.idxmax()
highest_avg_mpg_value = average_mpg_by_year.max()
print(f"\nThe year with the highest average mpg is: {year_highest_avg_mpg}")
print(f"The highest average mpg for that year was: {highest_avg_mpg_value:.2f}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = df_cars['mpg'].median()
print(f"\nMedian MPG of all cars: {median_mpg:.2f}")

above_median_mpg_cars = df_cars[df_cars['mpg'] > median_mpg].copy()

if above_median_mpg_cars.empty:
    print("\nNo cars found with MPG above the median. Cannot proceed with the analysis.")
else:
    print(f"Number of cars with MPG > {median_mpg:.2f}: {above_median_mpg_cars.shape[0]}")

    above_median_mpg_cars['hp_per_weight_ratio'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']
    max_hp_per_weight_ratio = above_median_mpg_cars['hp_per_weight_ratio'].max()
    print(f"Highest HP/Weight Ratio among above-median MPG cars: {max_hp_per_weight_ratio:.4f}")
    best_ratio_cars = above_median_mpg_cars[above_median_mpg_cars['hp_per_weight_ratio'] == max_hp_per_weight_ratio]

    print("\nCar(s) with the best horsepower-to-weight ratio among cars with above-median MPG:")
    for index, row in best_ratio_cars.iterrows():
        print(f"- {row['name']} (HP/Weight Ratio: {row['hp_per_weight_ratio']:.4f}, MPG: {row['mpg']})")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
try:
    df_cars = pd.read_csv('Cars.csv')
    print("Dataset 'Cars.csv' reloaded successfully for plotting.")
except FileNotFoundError:
    print("Error: 'Cars.csv' not found. Please verify the file name and path.")
    exit()

average_mpg_evolution = df_cars.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

print("\nAggregated data for plotting (first 5 rows):")
print(average_mpg_evolution.head())
plt.figure(figsize=(12, 7)) 
sns.lineplot(data=average_mpg_evolution, x='model_year', y='mpg', hue='origin', marker='o')
plt.title('Evolution of Average MPG Over Years by Car Origin', fontsize=16)
plt.xlabel('Model Year', fontsize=12)
plt.ylabel('Average MPG', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7) 
plt.xticks(average_mpg_evolution['model_year'].unique()) 
plt.legend(title='Origin') 
plt.tight_layout()
plt.show() 
```

### Create a Seaborn scatterplot (or PairGrid) where:

X = horsepower

Color by: origin

Add meaningful plot titles and axis titles.

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
try:
    df_cars = pd.read_csv('Cars.csv')
    print("Dataset 'Cars.csv' reloaded successfully for consistent car analysis.")
except FileNotFoundError:
    print("Error: 'Cars.csv' not found. Please verify the file name and path.")
    exit()

car_counts = df_cars['name'].value_counts()
multi_year_cars_names = car_counts[car_counts > 1].index

if multi_year_cars_names.empty:
    print("\nNo car models appeared in more than one model year.")
else:
    print(f"\nFound {len(multi_year_cars_names)} car models appearing in multiple years.")

    filtered_df = df_cars[df_cars['name'].isin(multi_year_cars_names)].copy()
    car_stats = filtered_df.groupby('name').agg(
        num_appearances=('name', 'count'), 
        avg_mpg=('mpg', 'mean'),          
        std_mpg=('mpg', 'std')             
    ).reset_index()
  
    car_stats['std_mpg'] = car_stats['std_mpg'].fillna(0) 

    consistent_cars = car_stats[car_stats['std_mpg'] < 1.0]
    consistent_cars = consistent_cars[consistent_cars['num_appearances'] >= 2]


    if consistent_cars.empty:
        print("\nNo consistent car models found with standard deviation of MPG < 1.0.")
    else:
        final_report = consistent_cars[['name', 'num_appearances', 'avg_mpg', 'std_mpg']].sort_values(
            by=['num_appearances', 'avg_mpg'],
            ascending=[False, False]
        )

        print("\n--- Consistent Car Models (Standard Deviation of MPG < 1.0) ---")
        print("Sorted by Number of Appearances (Descending), then Average MPG (Descending)")
        print("-" * 80)
        for index, row in final_report.iterrows():
            print(f"Model Name: {row['name']}")
            print(f"  Number of Appearances: {int(row['num_appearances'])}")
            print(f"  Average MPG: {row['avg_mpg']:.2f}")
            print(f"  Standard Deviation of MPG: {row['std_mpg']:.2f}")
            print("-" * 80)
```
